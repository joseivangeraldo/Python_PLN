```python
import spacy
texto = 'Os economistas do mercado financeiro mantiveram a estimativa de crescimento do Produto Interno Bruto (PIB) em 5,04% para 2021.O PIB é a soma de todos os bens e serviços produzidos no país e serve para medir a evolução da economia.Para 2022, o mercado baixou a previsão de alta do PIB de 1,63% para 1,57%. Foi a quarta redução seguida na estimativa de crescimento.Em setembro, o ministro da Economia, Paulo Guedes, admitiu que o "barulho político" dos últimos dias, resultado do discurso de caráter golpista do presidente Jair Bolsonaro nas manifestações de 7 de Setembro, poderia afetar o ritmo de crescimento da economia brasileira, gerando uma desaceleração.'
nlp = spacy.load("pt_core_news_lg")
doc = nlp(texto)
```

```python
tokens =[token for token in doc]
tokens
```

```python
###agora vamos recuperar o texto que esta dentro do token.Pegamos uma diretiva interna do spacy chamada orth
tokens =[token.orth_ for token in doc]
tokens ###agora as palavras são reornadas como strings individuais,e não como um objeto interno do pacote spacy
```

```python
##recuperar somente os tokens que forem caracteres
alpha_tokens =[token.orth_ for token in doc if token.is_alpha]
print(alpha_tokens)

alpha_digits =[token.orth_ for token in doc if token.is_digit]
print(alpha_digits)


puncts_digits =[token.orth_ for token in doc if token.is_punct]
print(puncts_digits)
```

```python
corpus = open('/content/drive/MyDrive/Arquivo_txt/corpus_teste.txt').read()
corpus
```

```python
import spacy
nlp = spacy.load('pt_core_news_lg')
doc = nlp(corpus)

tokens = [token.orth_ for token in doc]
tokens
```

```python
##recuperar somente os tokens que forem caracteres
alpha_tokens =[token.orth_ for token in doc if token.is_alpha]
print(alpha_tokens)

alpha_digits =[token.orth_ for token in doc if token.is_digit]
print(alpha_digits)


puncts_digits =[token.orth_ for token in doc if token.is_punct]
print(puncts_digits)
```

```python
lemmas = [token.lemma_ for token in doc if token.pos_ == 'VERB'] ###só funciona para verbo
lemmas
```

```python
pos =[(token.orth_,token.pos_) for token in doc]
pos
```

```python
##agora quero somente o texto seguido das informações morfologicas deste texto
morfologicas= [(token.orth_,token.morph) for token in doc]
morfologicas
```

```python
##gerdor de entidades nomeadas
entidades_nomeadas = list(doc.ents)
print(entidades_nomeadas)

###decobrir detalhes desta entidades
detalhes_entidades = [(entidade,entidade.label_) for entidade in doc.ents]
detalhes_entidades
```

```python
###exibir saida em uma forma html mais amigavel
html = spacy.displacy.render(doc ,style ='ent')
output_path = open('/content/drive/MyDrive/Arquivo_txt/entidades_nomeadas.html','w',encoding='utf-8')
output_path.write(html)
output_path.close()
```

```python
sintaxe = [(token.orth_,token.dep_) for token in doc]
sintaxe
```

```python
visualizar_sintaxe = spacy.displacy.render(doc ,style ='ent')
output_path = open('/content/drive/MyDrive/Arquivo_txt/analise_dependencia.svg','w',encoding='utf-8')
output_path.write(visualizar_sintaxe)
output_path.close()
```
