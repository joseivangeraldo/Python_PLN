```python
##FREQUENCIA DE PALAVRAS EM UM TEXTO
texto = "Com um passe de Eli Manning para Plaxico Burress a 39 segundos do fim, o New York Giants anotou o touchdown decisivo e derrubou o favorito New England Patriots por 17 a 14 neste domingo, em Glendale, no Super Bowl XLII."
tokens = nltk.word_tokenize(texto)
frequencia = nltk.FreqDist(tokens)
frequencia.most_common(10)  ##especifica as principais ou mais frequentes
```

```python
###vamos carregar o arquivo do corpus teste do google drive
corpus = open('/content/drive/MyDrive/Arquivo_txt/corpus_teste.txt').read()
print(corpus)
```

```python
##FREQUENCIA DE PALAVRAS EM UM TEXTO
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(corpus)

##a diferença do anterior é que não tem mais a virgula
frequencia = nltk.FreqDist(tokens)
frequencia.most_common()  ##especifica as principais ou mais frequentes
```

```python
##forma mais simples de transformar todas as palavras em minusculas
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*') ###imprimir letras
tokens = tokenizer.tokenize(corpus)

nova_lista =[]
for token in tokens:
  nova_lista.append(token.lower())
nova_lista

###agora vamos ver a frequencia com todas as palavras na mesma caixa
frequencia = nltk.FreqDist(nova_lista)
frequencia.most_common()
```

```python
###cria a lista destop words do portugues
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
##forma mais simples de transformar todas as palavras em minusculas
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*') ###imprimir letras
tokens = tokenizer.tokenize(corpus)

nova_lista =[]
for token in tokens:
  if token.lower() not in stopwords:
    nova_lista.append(token)

frequencia = nltk.FreqDist(nova_lista) 
frequencia.most_common()  
```
